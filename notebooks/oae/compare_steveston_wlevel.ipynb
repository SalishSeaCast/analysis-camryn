```python
import numpy as np
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
```

```python
WSC = pd.read_csv('/ocean/cstang/MOAD/analysis-camryn/notebooks/oae/data/08MH028_240730.csv')
WSC['Datetime'] = [pd.to_datetime(date) for date in WSC['Date (PST)']]

CHS = pd.read_csv('/ocean/cstang/MOAD/analysis-camryn/notebooks/oae/data/07607_240730.csv')
CHS['Datetime'] = [pd.to_datetime(date) for date in CHS['Date (PST)']]
CHS['Datetime']

mask = np.isin(CHS['Datetime'],WSC['Datetime'])
CHS_match = CHS[mask]

mask = np.isin(WSC['Datetime'],CHS['Datetime'])
WSC_match = WSC[mask]
```

```python
fig,ax = plt.subplots(figsize=(9,5))

ax.plot(CHS_match['Datetime'],CHS_match['observations (m)'],label='CHS')
ax.plot(WSC_match['Datetime'],WSC_match['Value (m)'],label='WSC')
ax.legend()
ax.set_ylabel('Water level (m)')
ax.set_title('Water level in June')
```
